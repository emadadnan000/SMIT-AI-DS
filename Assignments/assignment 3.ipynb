### Part 2: GroupBy Operations

```python
# Group by type and count the number of titles in each category
type_counts = df.groupby('type')['title'].count().reset_index()
type_counts
```

```python
# Group by country and find the top 5 countries with the most titles
top_countries = df['country'].value_counts().head(5)
print("Top 5 countries with the most titles:")
print(top_countries)
```

```python
# Group by rating and find average release year for each rating type
df_filtered_rating = df[~df['rating'].str.contains('min', na=False)]
rating_avg_year = df_filtered_rating.groupby('rating')['release_year'].mean().reset_index()
rating_avg_year.rename(columns={'release_year': 'average_release_year'}, inplace=True)
rating_avg_year['average_release_year'] = rating_avg_year['average_release_year'].astype(int)
rating_avg_year
```

```python
#  Group by release_year and show: total number of titles and count of movies and tv shows separately
release_year_summary = df.groupby('release_year').agg(
    total_titles=('title', 'count'),
    total_movies=('type', lambda x: (x == 'Movie').sum()),
    total_tv_shows=('type', lambda x: (x == 'TV Show').sum())
).reset_index()
release_year_summary
```

```python
# Create a new column `is_recent` True if `release_year` > 2015
df['is_recent'] = df['release_year'] > 2015
# Display the updated DataFrame with the new column
print(df[['title', 'release_year', 'is_recent']].head(10))
```

```python
# Group by country and count how many recent shows are there per country
recent_shows_by_country = df[df['is_recent']].groupby('country')['title'].count().reset_index()
recent_shows_by_country.rename(columns={'title': 'recent_shows_count'}, inplace=True)
# Display the recent shows count by country
print(recent_shows_by_country.sort_values(by='recent_shows_count', ascending=False))
```

```python
# Most common rating for TV Shows
tv_shows = df[df['type'] == 'TV Show']
most_common_tv_rating = tv_shows['rating'].mode()[0]
print("Most common rating for TV Shows:", most_common_tv_rating)
```

```python
# Country with the most content
most_content_country = df['country'].value_counts().idxmax()
print("Country with the most content:", most_content_country)
```

```python
# Compare number of titles before and after 2015
recent_count = (df['release_year'] > 2015).sum()
older_count = (df['release_year'] <= 2015).sum()
if recent_count > older_count:
    print("Most Netflix titles are recent (after 2015).")
else:
    print("Most Netflix titles are older (2015 or before).")
```

```python
# Average release year for Movies
movies = df[df['type'] == 'Movie']
avg_movie_year = int(movies['release_year'].mean())
print("Average release year for Netflix Movies:", avg_movie_year)
```

This suggests Netflix's catalog is modern and primarily targeted at adult audiences.
