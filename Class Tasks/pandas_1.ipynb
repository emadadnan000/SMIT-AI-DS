# Task 2

Create a DataFrame for 5 employees with Name, Department, Salary, and Experience. \
Add a column Bonus = 15% of salary. \
Filter employees with more than 3 years of experience. \
Sort employees based on Salary in descending order. 


```python
# Create a DataFrame for 5 employees with Name, Department, Salary, and Experience
employees = [
    {"Name": "Ali", "Department": "HR", "Salary": 60000, "Experience": 2},
    {"Name": "Sara", "Department": "Finance", "Salary": 75000, "Experience": 5},
    {"Name": "John", "Department": "IT", "Salary": 90000, "Experience": 4},
    {"Name": "Ayesha", "Department": "Marketing", "Salary": 65000, "Experience": 3},
    {"Name": "Bilal", "Department": "Sales", "Salary": 70000, "Experience": 6}
]

df_employees = pd.DataFrame(employees)
df_employees
```

```python
# Add a column Bonus = 15% of salary.
df_employees['Bonus'] = df_employees['Salary'] * 0.15
df_employees
```

```python
#Filter employees with more than 3 years of experience. 
experienced_employees = df_employees[df_employees['Experience'] > 3]
df_employees
```

```python
# Sort employees based on Salary in descending order. 
sorted_employees = df_employees.sort_values(by='Salary', ascending=False)
sorted_employees
```
