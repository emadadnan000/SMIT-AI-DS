```python
import numpy as np         
import matplotlib.pyplot as plt
```

## 1. Create a line plot of square numbers from 1â€“10.

```python
x = np.arange(1, 11)
y = x ** 2
plt.plot(x, y, marker ='o', linestyle='-')
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
```

## 2. Create a vertical and horizontal bar chart of student marks.

```python
x_ = np.array(["Maths", "Science", "English", "History", "Geography"])
y_ = np.array([85, 75, 65, 55, 50])

plt.bar(x_, y_, color='green', edgecolor='black')
plt.title('Bar Plot Example',color='blue',fontsize=20)
plt.xlabel('Subjects', fontsize=15, color='red')
plt.ylabel('Marks', fontsize=15, color='red')
```

```python
x_ = np.array(["Maths", "Science", "English", "History", "Geography"])
y_ = np.array([90, 80, 70, 60, 55])

plt.barh(x_, y_, color='green', edgecolor='black')
plt.title('Horizontal Bar Plot Example', color='blue', fontsize=20)
plt.xlabel('Marks', fontsize=15, color='red')
plt.ylabel('Subjects', fontsize=15, color='red')
```

## 3. Make a scatter plot using np.random.rand() values.

```python
a = np.random.rand(10)
b = np.random.rand(10)
plt.scatter(a, b, color='purple', marker='.')
plt.title('Scatter Plot Example', fontsize=20, color='orange')
plt.xlabel('X-axis', fontsize=15, color='blue')
plt.ylabel('Y-axis', fontsize=15, color='blue')
```

## 4. Create a histogram from 1000 normally distributed values.

```python
a_row = np.random.rand(1000)
plt.hist(a_row, bins=30, color='skyblue', edgecolor='black')
plt.title('Simple Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')  
```
