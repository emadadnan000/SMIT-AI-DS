1. Create a 4X2 integer array and Prints its attributes

The shape of an array.

Array dimensions.

The Length of each element of the array in bytes.

```python
array_1 = np.array([
    [1, 2],
    [3, 4],
    [5, 6],
    [7, 8]
])
print("Shape of the array:", array_1.shape)
print("Data type of the array:", array_1.dtype)
print("Number of dimensions of the array:", array_1.ndim)
print("Total number of elements in the array:", array_1.size)
print("Size of each element in bytes:", array_1.itemsize)
print("Total number of bytes consumed by the elements of the array:", array_1.nbytes)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import data
from skimage.color import rgb2gray
```

```python
# Load an example image from skimage
image = data.astronaut()
# Display the image using matplotlib
plt.imshow(image)
plt.title('Astronaut Image')
plt.axis('off')  # Hide the axes
plt.show()

```

```python
print(image.shape)
```

```python
# Converting the image to grayscale
gray_image = rgb2gray(image)
# Display the grayscale image using matplotlib
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Astronaut Image')
plt.axis('off')  # Hide the axes
plt.show()
```

```python
image_no_red = image.copy()
image_no_red[..., 0] = 0  # Set the red channel to 0

# Display the modified image
plt.imshow(image_no_red)
plt.title('Image with Red Pixels Turned Off')
plt.axis('off')  # Hide the axes
plt.show()
```

```python
image_no_green = image.copy()
image_no_green[..., 1] = 0  # Set the green channel to 0

# Display the modified image
plt.imshow(image_no_green)
plt.title('Image with Green Pixels Turned Off')
plt.axis('off')  # Hide the axes
plt.show()
```

```python
image_no_blue = image.copy()
image_no_blue[..., 2] = 0  # Set the blue channel to 0

# Display the modified image
plt.imshow(image_no_blue)
plt.title('Image with Blue Pixels Turned Off')
plt.axis('off')  # Hide the axes
plt.show()
```

```python
import matplotlib.pyplot as plt

from skimage import data
from skimage.color import rgb2gray

original = data.astronaut()
grayscale = rgb2gray(original)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()
ax[0].axis('off')
ax[0].imshow(original)
ax[0].set_title("Original")
ax[1].axis('off')
ax[1].imshow(grayscale, cmap=plt.cm.gray) # type: ignore
ax[1].set_title("Grayscale")

fig.tight_layout()
plt.show()
```
