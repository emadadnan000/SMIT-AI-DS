```python
import numpy as np         
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 3, 5, 7, 9])
y = np.array([2, 4, 6, 8, 10])
```

## Line Plot

```python
plt.plot(x, y, marker ='o', linestyle='-')
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
```

## Bar Plot

```python
x_ = np.array(["Maths", "Science", "English", "History", "Geography"])
y_ = np.array([85, 75, 65, 55, 50])

plt.bar(x_, y_, color='green', edgecolor='black')
plt.title('Bar Plot Example',color='blue',fontsize=20)
plt.xlabel('Subjects', fontsize=15, color='red')
plt.ylabel('Marks', fontsize=15, color='red')
```

## Scatter Plot

```python
a = np.arange(50)
b = np.arange(50)
plt.scatter(a, b, color='purple', marker='.')
plt.title('Scatter Plot Example', fontsize=20, color='orange')
plt.xlabel('X-axis', fontsize=15, color='blue')
plt.ylabel('Y-axis', fontsize=15, color='blue')
```

## Scatter Plot

```python
hours_studied = np.arange(1, 8)
marks_scored = np.array([50, 60, 70, 80, 90, 95, 100])
plt.scatter(hours_studied, marks_scored, color='red', marker='x')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Scored')
```

## Histogram


```python
a_row = np.random.rand(10)
plt.hist(a_row, bins=5, color='skyblue', edgecolor='black')
plt.title('Simple Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')  # Frequency of values in array 'b'
```

## Box Plot

```python
# Example: Box plot of daily step counts for a week
daily_steps = [7000, 8500, 9000, 7500, 8000, 9500, 10000]
plt.boxplot(daily_steps, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Box Plot of Daily Step Counts')
plt.ylabel('Steps')
plt.xticks([1], ['Week'])
plt.show()
```

## Pie Chart

```python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 10, 25, 25]
colors = ['gold', 'lightcoral', 'lightskyblue', 'yellowgreen']
explode= (0.3, 0.1, 0, 0)  

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True)
```
