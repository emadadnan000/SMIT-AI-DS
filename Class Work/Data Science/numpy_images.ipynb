### Use all Image numpy methods


```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import data
from skimage.color import rgb2gray
```

```python
# Load an example image from skimage
image = data.coffee()
# Display the image using matplotlib
plt.imshow(image)
plt.title('Coffee Image')
print("The shape of the image is: ", image.shape)
plt.axis('off')  # Hide the axes
plt.show()
```

```python
image_no_red = image.copy()
image_no_red[..., 0] = 0  # Set the red channel to 0

# Display the modified image
plt.imshow(image_no_red)
plt.title('Image with Red Pixels Turned Off')
plt.axis('off')  # Hide the axes
plt.show()
```

```python
image_no_green = image.copy()
image_no_green[..., 1] = 0  # Set the green channel to 0

# Display the modified image
plt.imshow(image_no_green)
plt.title('Image with Green Pixels Turned Off')
plt.axis('off')  # Hide the axes
plt.show()
```

```python
image_no_blue = image.copy()
image_no_blue[..., 2] = 0  # Set the blue channel to 0

# Display the modified image
plt.imshow(image_no_blue)
plt.title('Image with Blue Pixels Turned Off')
plt.axis('off')  # Hide the axes
plt.show()
```

```python

# Crop the image using slicing
cropped_image = image[100:400, 200:400]

# Display the cropped image
plt.imshow(cropped_image)
plt.title('Cropped Image')
plt.axis('off')  # Hide the axes
plt.show()
```

```python

flipped_vertically = np.flip(image, axis=0)
plt.imshow(flipped_vertically)
plt.title('Flipped Vertically')
plt.axis('off')
plt.show()
```

```python
flipped_horizontally = np.flip(image, axis=1)  # Flip along the horizontal axis
plt.imshow(flipped_horizontally)
plt.title('Flipped Horizontally')
plt.axis('off')  # Hide the axes
plt.show()
```

```python
flipped_horizontally = image[:, ::-1, :]  # Reverse the columns

# Display the flipped image
plt.imshow(flipped_horizontally)
plt.title('Flipped Horizontally')
plt.axis('off')  # Hide the axes
plt.show()
```

```python
flipped_vertically = image[::-1, :, :]  # Reverse the rows

# Display the flipped image
plt.imshow(flipped_vertically)
plt.title('Flipped Vertically')
plt.axis('off')  # Hide the axes
plt.show()
```

```python

original = data.coffee()
grayscale = rgb2gray(original)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()
ax[0].axis('off')
ax[0].imshow(original)
ax[0].set_title("Original")
ax[1].axis('off')
ax[1].imshow(grayscale, cmap=plt.cm.gray)
ax[1].set_title("Grayscale")

fig.tight_layout()
plt.show()
```
