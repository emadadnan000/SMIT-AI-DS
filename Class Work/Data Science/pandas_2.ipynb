```python
import pandas as pd
import seaborn as sns
```

```python
# load the tips dataset
df = sns.load_dataset("tips")
df.head()
```

```python
print("First 5 rows of the DataFrame:\n")
print(df.head())

print("\nLast 5 rows of the DataFrame:\n")
print(df.tail())

print("\nShape of the DataFrame (rows, columns):\n")
print(df.shape)

print("\nColumn names in the DataFrame:\n")
print(df.columns)

```

```python
print("\nDataFrame info:\n")
df.info()

print("\nCheck for missing values (True if missing):\n")
print(df.isnull())

```

```python

print("\nDataFrame after dropping rows with missing values:\n")
df.dropna()
```

```python

print("\nDataFrame sorted by total_bill in descending order:\n")
df.sort_values(by='total_bill', ascending=False)
```

```python
#new columns: bill_per_person = total_bill / size & is_weekend = True if day is Sat/Sun 
df['bill_per_person'] = df['total_bill'] / df['size']
df['is_weekend'] = df['day'].isin(['Sat', 'Sun'])
df
```

```python
# Group by day and analyze tip and bill
grouped = df.groupby('day').agg({'total_bill': 'mean', 'tip': 'mean'}).reset_index()
print("\nGrouped DataFrame by day with mean total_bill and bill:\n")
grouped
```
