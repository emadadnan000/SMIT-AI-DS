## üß† 1. Introduction

In deep learning, a neural network learns to make predictions ‚Äî but two important components control **how it learns** and **how it makes decisions**:

1. **Activation Functions** ‚Üí decide *how a neuron activates (fires)*  
2. **Loss Functions** ‚Üí measure *how wrong the model‚Äôs predictions are*  

Together, they are the **heart and brain** of a neural network.



## ‚ö° 2. Activation Function (Simple Explanation)

An **activation function** decides **whether a neuron should be active or not** based on its input.

Think of it like a **gate** that controls what information should pass through.



### üîπ Example (Real-Life Analogy)
Imagine a person deciding whether to buy a product:
- If the price is **too high**, they say **NO** (inactive)
- If it‚Äôs **reasonable**, they say **YES** (active)

This ‚Äúdecision behavior‚Äù is what an activation function does in a neural network.



### üîπ Why Do We Need Activation Functions?

Without activation functions, a neural network would just be a **linear system** ‚Äî no matter how many layers it has, it can only learn straight-line relationships.

Activation functions allow the network to learn **complex, non-linear patterns** like:
- Images
- Speech
- Emotions
- Handwriting



## üîπ 3. Common Activation Functions

| Activation Function | Formula (short) | Used In | Description |
|----------------------|----------------|----------|--------------|
| **Sigmoid** | f(x) = 1 / (1 + e^(-x)) | Output layer (binary classification) | Converts value to 0‚Äì1 (probability) |
| **ReLU (Rectified Linear Unit)** | f(x) = max(0, x) | Hidden layers (most common) | Keeps positive values, sets negatives to 0 |
| **Tanh** | f(x) = (e^x - e^-x) / (e^x + e^-x) | Hidden layers | Outputs between -1 and 1 |
| **Softmax** | f(x_i) = e^(x_i) / Œ£ e^(x_j) | Output layer (multi-class classification) | Converts outputs into class probabilities |


```python

import numpy as np
import matplotlib.pyplot as plt

# Generate some data
x = np.linspace(-10, 10, 100)

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Plot
plt.figure(figsize=(8,5))
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, tanh(x), label='Tanh')
plt.title("Activation Functions")
plt.legend()
plt.grid(True)
plt.show()
```


### üß© Explanation:
- We created a range of values from -10 to 10.  
- Defined **Sigmoid**, **ReLU**, and **Tanh** functions.  
- Plotted them to **see how each one transforms input values**.

You‚Äôll notice:
- **Sigmoid** squashes everything between 0 and 1.  
- **ReLU** keeps positives, removes negatives.  
- **Tanh** squashes between -1 and 1.



## ‚ù§Ô∏è 6. Loss Function (Simple Explanation)

The **loss function** tells the model **how wrong it is** after each prediction.

üëâ It‚Äôs like a **report card** for your model after every exam (epoch).

The **smaller the loss**, the better your model is learning.



### üîπ Example
If real answer = 100  
and model predicted = 90  

Then the **loss (error)** = 10 (the difference).

The model then tries to adjust itself to make this loss smaller next time.



## üîπ 7. Types of Loss Functions

| Type | Common Loss Function | Used For | Formula (short) |
|------|----------------------|-----------|------------------|
| **Regression** | Mean Squared Error (MSE) | Predicting numbers (like price, temperature) | (1/n) Œ£(y - ≈∑)^2 |
| **Binary Classification** | Binary Cross-Entropy | Yes/No problems | -[y log(≈∑) + (1-y) log(1-≈∑)] |
| **Multi-Class Classification** | Categorical Cross-Entropy | Multi-class problems | -Œ£ y_i log(≈∑_i) |


```python

import numpy as np

# Actual and predicted values
y_true = np.array([3.0, 4.5, 2.0, 6.0])
y_pred = np.array([2.5, 5.0, 2.2, 5.5])

# Mean Squared Error
mse_loss = np.mean((y_true - y_pred) ** 2)
print("Mean Squared Error:", mse_loss)
```


### üß© Explanation:
- Compare actual and predicted values.  
- Subtract to find errors.  
- Square them to remove negatives.  
- Take the mean ‚Üí final loss value.  

If the loss = **0**, model predictions are perfect!


```python

# Actual and predicted probabilities
y_true = np.array([1, 0, 1, 0])
y_pred = np.array([0.9, 0.1, 0.8, 0.2])

# Binary Cross Entropy Loss
bce_loss = -np.mean(y_true*np.log(y_pred) + (1 - y_true)*np.log(1 - y_pred))
print("Binary Cross-Entropy Loss:", bce_loss)
```


### üß© Explanation:
- When the model predicts probabilities close to actual values ‚Üí loss is low.  
- When it‚Äôs far ‚Üí loss is high.  
- The log function heavily penalizes confident wrong predictions.



## ‚öôÔ∏è 10. When to Use Which Function

| Problem Type | Activation Function | Loss Function |
|---------------|--------------------|---------------|
| **Regression** | Linear / ReLU | MSE |
| **Binary Classification** | Sigmoid | Binary Cross-Entropy |
| **Multi-Class Classification** | Softmax | Categorical Cross-Entropy |
