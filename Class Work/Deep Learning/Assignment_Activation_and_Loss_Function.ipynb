```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some data
x = np.linspace(-10, 10, 100)

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Plot activation functions
plt.figure(figsize=(10, 6))

plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, tanh(x), label='Tanh')

plt.title('Activation Functions')
plt.xlabel('x')
plt.ylabel('Output')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create a small NumPy dataset for regression
np.random.seed(42)  # for reproducibility
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Calculate Mean Absolute Error (MAE) manually
def mean_absolute_error(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

# Let's make some random predictions for demonstration
y_pred = 5 + 2 * X + np.random.randn(100, 1)

mae = mean_absolute_error(y, y_pred)
print(f"Mean Absolute Error (MAE): {mae}")
```

```python
# Write code to compute Binary Cross-Entropy Loss for random predictions.
def binary_cross_entropy(y_true, y_pred):
    # Ensure predictions are not exactly 0 or 1 to avoid log(0)
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

# Generate some random true labels (0 or 1) and predictions (between 0 and 1)
np.random.seed(42)
y_true = np.random.randint(0, 2, 100)
y_pred = np.random.rand(100)

bce_loss = binary_cross_entropy(y_true, y_pred)
print(f"Binary Cross-Entropy Loss: {bce_loss}")
```

```python
from sklearn.datasets import make_classification

# Generate a synthetic classification dataset
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

def create_model(activation):
    """Creates a simple Keras model with a specified activation function."""
    model = Sequential([
        Dense(10, activation=activation, input_shape=(X.shape[1],)),
        Dense(1, activation='sigmoid')
    ])
    return model

# Create models with different activation functions
model_relu = create_model('relu')
model_sigmoid = create_model('sigmoid')
model_tanh = create_model('tanh')

print("ReLU Model Summary:")
model_relu.summary()

print("\nSigmoid Model Summary:")
model_sigmoid.summary()

print("\nTanh Model Summary:")
model_tanh.summary()
```

```python
# Compile models
model_relu.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_sigmoid.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_tanh.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

print("Models compiled successfully.")
```

```python
# Train models
epochs = 50
batch_size = 32

history_relu = model_relu.fit(X, y, epochs=epochs, batch_size=batch_size, verbose=0)
history_sigmoid = model_sigmoid.fit(X, y, epochs=epochs, batch_size=batch_size, verbose=0)
history_tanh = model_tanh.fit(X, y, epochs=epochs, batch_size=batch_size, verbose=0)

print("Models trained successfully.")
```

```python
# Evaluate models
evaluation_relu = model_relu.evaluate(X, y, verbose=0)
evaluation_sigmoid = model_sigmoid.evaluate(X, y, verbose=0)
evaluation_tanh = model_tanh.evaluate(X, y, verbose=0)

# Print evaluation results
print(f"ReLU Model - Loss: {evaluation_relu[0]:.4f}, Accuracy: {evaluation_relu[1]:.4f}")
print(f"Sigmoid Model - Loss: {evaluation_sigmoid[0]:.4f}, Accuracy: {evaluation_sigmoid[1]:.4f}")
print(f"Tanh Model - Loss: {evaluation_tanh[0]:.4f}, Accuracy: {evaluation_tanh[1]:.4f}")
```

```python
# Compare accuracies
accuracy_relu = evaluation_relu[1]
accuracy_sigmoid = evaluation_sigmoid[1]
accuracy_tanh = evaluation_tanh[1]

print("\n--- Activation Function Performance Comparison ---")
print(f"ReLU Model Accuracy: {accuracy_relu:.4f}")
print(f"Sigmoid Model Accuracy: {accuracy_sigmoid:.4f}")
print(f"Tanh Model Accuracy: {accuracy_tanh:.4f}")

# Identify best and worst
accuracies = {
    'ReLU': accuracy_relu,
    'Sigmoid': accuracy_sigmoid,
    'Tanh': accuracy_tanh
}

best_activation = max(accuracies, key=accuracies.get)
worst_activation = min(accuracies, key=accuracies.get)

print(f"\nBest performing activation function: {best_activation} (Accuracy: {accuracies[best_activation]:.4f})")
print(f"Worst performing activation function: {worst_activation} (Accuracy: {accuracies[worst_activation]:.4f})")

# Briefly describe differences
print("\nObserved differences:")
print(f"- The Tanh activation function achieved the highest accuracy ({accuracy_tanh:.4f}).")
print(f"- The Sigmoid activation function achieved the lowest accuracy ({accuracy_sigmoid:.4f}).")
print(f"- The ReLU activation function performed in between Tanh and Sigmoid ({accuracy_relu:.4f}).")
```

```python
from sklearn.datasets import make_classification

# Generate a synthetic classification dataset
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Instantiate a Sequential model
model_relu_sigmoid = Sequential([
    # First hidden layer with ReLU
    Dense(10, activation='relu', input_shape=(X.shape[1],)),
    # Second hidden layer with ReLU
    Dense(10, activation='relu'),
    # Output layer with Sigmoid for binary classification
    Dense(1, activation='sigmoid')
])

# Print the model summary
model_relu_sigmoid.summary()
```

```python
# Compile the model
model_relu_sigmoid.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

print("Model compiled successfully.")
```

```python
# Train the model
epochs = 100  # You can adjust the number of epochs
batch_size = 32 # You can adjust the batch size

history_relu_sigmoid = model_relu_sigmoid.fit(X, y, epochs=epochs, batch_size=batch_size, verbose=0)

print("Model trained successfully.")
```

```python
# Access the loss values from the training history
loss_values = history_relu_sigmoid.history['loss']

# Create a plot of the loss values
plt.figure(figsize=(10, 6))
plt.plot(loss_values)

# Add title and labels
plt.title('Model Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')

# Add a grid
plt.grid(True)

# Display the plot
plt.show()
```
