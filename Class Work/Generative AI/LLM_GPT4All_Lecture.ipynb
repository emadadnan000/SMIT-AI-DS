## 2️⃣ Import and Initialize Model

```python
from gpt4all import GPT4All

# Load the local model
# model = GPT4All("mistral-7b-instruct-v0.1.Q4_0.gguf")
model = GPT4All("Phi-3-mini-4k-instruct.Q4_0.gguf")
```

## 3️⃣ Example 1: Ask a Question

```python
response = model.generate("Explain Python loops in simple words for a beginner.")
print("Example 1 - Explanation of Python loops:")
print(response)
```

## 4️⃣ Example 2: Summarize a Paragraph

```python
text = """
Python is a high-level, interpreted programming language.
It is widely used for web development, data analysis, artificial intelligence, and more.
"""

summary = model.generate(f"Summarize this text in one sentence: {text}")
print("\nExample 2 - Summary of text:")
print(summary)
```

## 5️⃣ Example 3: Mini Interactive Chatbot

```python
print("\nMini Chatbot (type 'exit' to stop)")
while True:
    question = input("You: ")
    if question.lower() == "exit":
        break
    answer = model.generate(question)
    print("Bot:", answer)
```

## 6️⃣ Example 4: Simple "Chain" Example (Summarize + Translate)

```python
paragraph = "Python is a high-level programming language used for AI, web apps, and data analysis."

# Step 1: Summarize
summary = model.generate(f"Summarize this text: {paragraph}")

# Step 2: Translate
translation = model.generate(f"Translate this summary to French: {summary}")

print("\nExample 4 - Chain Example")
print("Summary:", summary)
print("Translation:", translation)
```
