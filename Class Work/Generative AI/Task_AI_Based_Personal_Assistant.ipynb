```python
from gpt4all import GPT4All
# Download Gemma models from the GPT4All website (e.g., 'https://gpt4all.io/models/gguf/gemma-2b-it-q4_k_m.gguf')
# and place them in the same directory as your notebook or provide the full path.

# model = GPT4All("orca-mini-3b-gguf2-q4_0.gguf") # include .bin
# model = GPT4All("Meta-Llama-3-8B-Instruct.Q4_0.gguf") # Your previous model (commented out)
model = GPT4All("Phi-3-mini-4k-instruct.Q4_0.gguf") # Example for a Gemma 2B model - replace with your desired Gemma 3 model filename
```

## üëâ Step 2: Generate a Personalized Day Plan

Generate a personalized schedule that includes:

Study plan

Break suggestions

Time allocation

Task priority (High/Medium/Low)

```python
name = input("Enter your nameWN: ")
print('Enter 3 goals of yours: ')
goal1 = input("Goal 1: ")
goal2 = input("Goal 2: ")
goal3 = input("Goal 3: ")
hrs = input("how many study hours you have available: ")
topic = input("what is the topic you want to study: ")
```

```python
response = model.generate(f"Create a personalized productivity plan for a student named {name}.His goals: {goal1}, {goal2}, {goal3}.Available study hours: {hrs}.Focus topic: {topic}.Include time blocks, study suggestions, and breaks.")
print(response)
```


## üëâ Step 3: Build a Chain (Two-Step Processing)

1. First step:
Generate a detailed plan.


2. Second step:
Ask the model to summarize the plan in bullet points.



Summary Prompt Example:

Summarize the above plan into 5 bullet points.

```python
summary_prompt = f"Summarize the following plan into 5 bullet points:\n\n{response}"
summary_response = model.generate(summary_prompt)
print(summary_response)
```
