# Introduction to Loops

Sometimes a programmer wants to execute a group of statements a certain number of times. This can be done using loops. Based on this loops are further classified into following main types;

1. For Loops
2. While Loops

```python
name = 'Abcdefg'
for i in name:
  print(i)
```

```python
task = "Hello World"

for i in task:
  print(i)
```

### range():
What if we do not want to iterate over a sequence? What if we want to use for loop for a specific number of times?

Here, we can use the range() function.

```python
for k in range(0, 5):
    print(k , end = ",")
```

```python
for i in range(1,11):
  print(5  ,"x" , i , "="  , 5 * i)
```

```python
table_no = int(input("Enter Table no: "))
start_range = int(input("Enter starting range: "))
end_range = int(input("Enter ending range: "))

for i in range(start_range , end_range + 1):
  print(table_no , "x" , i , "=" , table_no * i)
```

```python
for i in range(1 , 11):
  if i == 5:
    break
  else:
    print(i)
```

```python
for i in range(1,51):
  if i == 45:
    continue
  else:
    print(i)
```

Here, we can see that the loop starts from 0 by default and increments at each iteration.

But we can also loop over a specific range.

```python
table_no = int(input("Enter a number: "))
starting_range = int(input("Define Starting Range: "))
ending_range = int(input("Define Ending Range: "))


for i in range(starting_range , ending_range + 1):
  print(table_no , "x" , i , "=" , table_no*i)
```

```python
for i in range(1,101):
  if i == 50:
    continue
  else:
    print(i)
```

```python
for i in range(1,12):
  if(i == 5):
    break
  else:
    print(i)
```

```python
a = "Hello World!"

for i in a:
  if i == "o":
    print(a.index(i))
    break
  else:
    print(False)
```

```python
a = "In Babylonian astronomy"

for i in a:
  if i == "B":
    print("The Index number of i: " , a.index(i))
    break
  else:
    print(i)
```
