## ðŸŒ³ 1. What is a Decision Tree?
A **Decision Tree** is a supervised machine learning algorithm used for **classification** and **regression** tasks.  

- It works like a **flowchart of questions and answers**.  
- At each step, the algorithm asks a **question** about the data (e.g., "Is petal length < 2.5 cm?").  
- Based on the answer (Yes/No), it splits the data into groups.  
- This continues until the tree reaches a decision (final class or value).  

ðŸ‘‰ Think of it like playing **20 Questions** where each question narrows down the possibilities.



## ðŸ§  2. Key Concepts (Simple Explanation)

- **Node** â†’ A point where a **question** is asked.  
- **Root Node** â†’ The **first question** in the tree (best split).  
- **Leaf Node** â†’ The final decision (e.g., "This flower is Setosa").  
- **Splitting Criteria** â†’ The algorithm decides which question to ask using:
  - **Gini Impurity** (default)  
  - **Entropy** (Information Gain)  

ðŸ‘‰ Goal: Make groups as **pure** as possible.



## ðŸ“ 3. Example (Iris Dataset)
The **Iris dataset** has features:  
- Sepal Length, Sepal Width, Petal Length, Petal Width  
- Target: Type of flower (Setosa, Versicolor, Virginica)


### ðŸ“Œ Step 1: Import Libraries

````python

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
````

### ðŸ“Œ Step 2: Load Dataset

````python

iris = load_iris()
X = iris.data
y = iris.target

print("Feature names:", iris.feature_names)
print("Target names:", iris.target_names)
````

### ðŸ“Œ Step 3: Split Data

````python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print("Training size:", len(X_train))
print("Testing size:", len(X_test))
````

### ðŸ“Œ Step 4: Train Decision Tree

````python

clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X_train, y_train)
````

### ðŸ“Œ Step 5: Evaluate Model

````python

accuracy = clf.score(X_test, y_test)
print("Accuracy:", accuracy)
````

### ðŸ“Œ Step 6: Visualize Tree

````python

plt.figure(figsize=(12,8))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
````
