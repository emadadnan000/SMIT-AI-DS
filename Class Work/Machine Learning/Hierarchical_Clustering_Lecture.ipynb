## ðŸ§  1. Introduction to Hierarchical Clustering

Hierarchical Clustering is an **unsupervised learning algorithm** used to group similar data points into clusters.  
Unlike K-Means, it **does not require specifying the number of clusters (K)** beforehand. Instead, it builds a **tree-like structure** of clusters called a **dendrogram**.

### ðŸ§© Key Idea:
- Hierarchical clustering works by **either merging smaller clusters into larger ones (Agglomerative)** or **dividing larger clusters into smaller ones (Divisive)**.

### ðŸ”¹ Types of Hierarchical Clustering:
1. **Agglomerative (Bottom-Up):** Each data point starts as its own cluster, and pairs of clusters are merged step by step.  
2. **Divisive (Top-Down):** All points start in one cluster, and the cluster splits recursively.

ðŸ‘‰ The **Agglomerative approach** is most commonly used.

## ðŸ§® 3. General Example: Grouping Students by Test Scores

Letâ€™s take a **simple and intuitive example** to understand how hierarchical clustering works.

### Example: Studentsâ€™ Math Scores

| Student | Score |
|----------|--------|
| A | 35 |
| B | 40 |
| C | 42 |
| D | 80 |
| E | 83 |
| F | 85 |

We want to **group students with similar performance**.

1. **Each Student as a Cluster:**  
   Initially, every student is treated as a separate cluster.

2. **Compute Distances:**  
   Distances between scores:  
   - Aâ€“B = 5  
   - Bâ€“C = 2  
   - Dâ€“E = 3  
   - Eâ€“F = 2  

   So, A-B-C form one group (low scorers) and D-E-F form another (high scorers).

3. **Merge Closest Clusters:**  
   Merge (B, C), then (A) joins them â†’ Low-score cluster.  
   Merge (E, F), then (D) joins them â†’ High-score cluster.

4. **Dendrogram:**  
   Cutting the dendrogram at the right height gives two clusters â€” low and high scorers.

## ðŸ’» 4. Practical Implementation (Simple Example)

### Step 1: Import Libraries


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

### Step 2: Create Simple Data

```python
X = np.array([
    [1, 2],
    [1.5, 1.8],
    [5, 8],
    [8, 8],
    [1, 0.6],
    [9, 11]
])
```

```python
plt.scatter(X[:,0], X[:,1], color='blue')
plt.title('Sample Data Points')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

### Step 3: Build a Dendrogram

```python
plt.figure(figsize=(8, 6))
dendrogram = sch.dendrogram(sch.linkage(X, method='average'))
plt.title('Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.show()
```

### Step 4: Train Hierarchical Clustering Model

```python
hc = AgglomerativeClustering(n_clusters=2, linkage='ward')
y_hc = hc.fit_predict(X)
print('Cluster Labels:', y_hc)
```

### Step 5: Visualize the Clusters

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s=100, c='blue', label='Cluster 2')
plt.title('Hierarchical Clustering Results')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```
