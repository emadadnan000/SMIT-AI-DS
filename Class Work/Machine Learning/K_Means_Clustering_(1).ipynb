## 1. Introduction to K-Means Clustering

- **Clustering**: grouping similar data points without labels (unsupervised learning).
- **K-Means**: partitions data into **K** clusters by iteratively assigning points to nearest centroids and updating centroids as cluster means.

**Algorithm (brief):**
1. Choose K
2. Initialize K centroids
3. Assign points to nearest centroid
4. Recompute centroids
5. Repeat until convergence

**Notes:** Works best with numeric features; scale features when ranges differ.

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Note: In Jupyter, use `%matplotlib inline` if desired.
```

**Explanation:** We import `numpy`, `pandas`, and `matplotlib` for data handling and plotting, and `KMeans` from `sklearn` for clustering.

```python
# Create sample customer dataset (Age & Annual Income)
data = {
    'Age': [22, 25, 47, 52, 46, 56, 55, 60, 62, 61, 23, 24, 26, 48, 50],
    'Annual_Income': [25000, 27000, 80000, 85000, 81000, 87000, 89000, 91000, 92000, 94000, 26000, 28000, 29000, 83000, 86000]
}

customers = pd.DataFrame(data)
customers.head()
```

**Explanation:** We created a small custom dataset of 15 customers with `Age` and `Annual_Income`. This is simple and suitable for demonstrating K-Means clustering on two features.

```python
# Plot raw customer data
plt.figure(figsize=(6,4))
plt.scatter(customers['Age'], customers['Annual_Income'])
plt.xlabel('Age')
plt.ylabel('Annual Income')
plt.title('Customers: Age vs Annual Income (raw)')
plt.show()
```

**Explanation:** This scatter plot shows Age vs Annual Income. Visual inspection may reveal natural groupings (e.g., younger low-income vs older high-income).

```python
# Apply K-Means with k=2
k = 2
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
customers['Cluster'] = kmeans.fit_predict(customers)

customers
```

**Explanation:** We set `k=2` and fit K-Means to the two features. The `Cluster` column shows each sample's assigned cluster (0 or 1).

```python
# Visualize the clusters
plt.figure(figsize=(6,4))
plt.scatter(customers['Age'], customers['Annual_Income'], c=customers['Cluster'], cmap='rainbow')
plt.xlabel('Age')
plt.ylabel('Annual Income')
plt.title('K-Means Clusters (k=2)')
# Plot centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:,0], centroids[:,1], marker='X', s=200)
plt.show()
```

**Explanation:** The plot colors each point by cluster assignment. Centroids are shown as large X markers. This helps visually confirm cluster separation.

```python
# Create a sample students dataset (Study Hours & Sleep Hours)
students_data = {
    'Study_Hours': [2, 3, 4, 5, 6, 8, 9, 10, 11, 12],
    'Sleep_Hours': [9, 8, 7, 7, 6, 5, 5, 4, 4, 3]
}

students = pd.DataFrame(students_data)
students
```

**Explanation:** We created a simple dataset of 10 students with `Study_Hours` and `Sleep_Hours`. This will be used for the class task demonstration.

## Class Task (with solution)

**Task:** Use the students dataset and apply K-Means with `k=2`. Visualize the clusters and show the cluster assignments.

Below is the solution code.

```python
# Class Task Solution
k = 2
kmeans_students = KMeans(n_clusters=k, random_state=42, n_init=10)
students['Cluster'] = kmeans_students.fit_predict(students)

# Plot
plt.figure(figsize=(6,4))
plt.scatter(students['Study_Hours'], students['Sleep_Hours'], c=students['Cluster'], cmap='viridis')
plt.xlabel('Study Hours')
plt.ylabel('Sleep Hours')
plt.title('Student Clusters (k=2)')
centroids_s = kmeans_students.cluster_centers_
plt.scatter(centroids_s[:,0], centroids_s[:,1], marker='X', s=150)
plt.show()

students
```

**Explanation (solution):**
- We fit K-Means with `k=2` on study vs sleep hours.
- The plot shows two clusters; typically one cluster groups high-study/low-sleep students and the other low-study/high-sleep.
- The DataFrame shows each student's cluster label.

## Student Task (without solution)  Practice

**Problem:**
1. Create a DataFrame of 1520 mobile users with two features: `Screen_Time` (hours per day) and `Battery_Usage` (percentage).
2. Apply K-Means with `k=3`.
3. Plot the clusters with distinct colors and mark centroids.
4. Write 1-2 sentences interpreting each cluster (e.g., heavy users, moderate users, light users).

*Try solving this in a new cell below.*

```python
# 1. Create a DataFrame of mobile users
mobile_data = {
    'Screen_Time': [2.5, 3.0, 4.2, 5.5, 6.0, 7.8, 8.5, 9.0, 2.0, 3.5, 4.8, 6.2, 7.0, 8.0, 9.5, 5.0, 3.2, 6.8, 7.5, 2.8],
    'Battery_Usage': [30, 35, 50, 65, 70, 85, 90, 95, 25, 40, 55, 75, 80, 88, 98, 60, 38, 78, 85, 32]
}
users = pd.DataFrame(mobile_data)

# 2. Apply K-Means with k=3
k = 3
kmeans_users = KMeans(n_clusters=k, random_state=42, n_init=10)
users['Cluster'] = kmeans_users.fit_predict(users)

# 3. Plot the clusters and centroids
plt.figure(figsize=(7,5))
plt.scatter(users['Screen_Time'], users['Battery_Usage'], c=users['Cluster'], cmap='Set1', s=60)
centroids = kmeans_users.cluster_centers_
plt.scatter(centroids[:,0], centroids[:,1], marker='X', s=200, c='black', label='Centroids')
plt.xlabel('Screen Time (hours/day)')
plt.ylabel('Battery Usage (%)')
plt.title('Mobile Users Clustering (k=3)')
plt.legend()
plt.grid(True)
plt.show()

users
```

**Explanation:**

- We created a DataFrame of 20 mobile users with two features: `Screen_Time` (hours per day) and `Battery_Usage` (percentage).
- K-Means clustering with `k=3` groups users into three clusters based on their usage patterns.
- The scatter plot colors each user by cluster and marks centroids with black Xs.
- You can interpret the clusters as follows:
    - One cluster likely represents heavy users (high screen time, high battery usage).
    - Another cluster may represent moderate users.
    - The third cluster likely groups light users (low screen time, low battery usage).
- Check the DataFrame to see each user's assigned cluster.
