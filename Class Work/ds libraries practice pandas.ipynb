### DataFrame


```python
data = {
    'Name': ['John', 'Alice', 'Bob'],
    'Age': [14, 15, 16],
    'Marks': [87, 68, 95]
}
df = pd.DataFrame(data)
df
```

```python
# Create from list of dictionaries
data = [
    {'Name': 'John', 'Age': 14, 'Marks': 87},
    {'Name': 'Alice', 'Age': 15, 'Marks': 68},
    {'Name': 'Bob', 'Age': 16, 'Marks': 95}
]
df2 = pd.DataFrame(data)
df2
```

```python
# Accessing Columns
print(df['Name'])  # Access by column name
print(df[['Name', 'Marks']])  # Access by column name

```

```python
print("Age")
print(df.Age)  # Access by attribute
```

```python
print(df.loc[0])  # Access first row
```

```python
print(df.iloc[1])  # Access first row by position
```

```python
# Adding a new column
df['Grade'] = ['A', 'B', 'A+']
df
```

```python
print(df[df['Scores'] > 80])  # Filter rows based on condition
```

```python
df.describe()
```

```python
# Renaming a column
df.rename(columns={'Marks': 'Scores'}, inplace=True)
df
```

```python
df.drop(columns=['Grade'], axis = 1, inplace=False)  # Drop a column
```

```python
# Sort by column
df.sort_values(by='Scores', ascending=True, inplace=False)
```

```python
df['Passes'] = df['Scores'] > 70  # Create a boolean column
print(df)
```

```python
#import seaborn as sns
df = sns.load_dataset('titanic')
df_clean = df.drop(['deck'], axis = 1)
df_clean['age'] = df_clean['age'].astype('float')
df_clean = df_clean.drop(['embark_town'], axis = 1)
df_clean.rename(columns={'sibsp': 'siblings_spouses_abroad'}, inplace=True)  # Rename a column
df_clean.rename(columns={'parch': 'parents_children'}, inplace=True)  # Rename a column
df_clean
```

```python
print(df_clean.isnull().sum())  # Check for missing values
```

```python
df_clean['age'].fillna(df_clean['age'].median())  # Fill missing values in 'age' with the median
df_clean.dropna(subset=['embarked'], inplace=True)  # Drop rows where 'embarked' is NaN
df_clean
```
