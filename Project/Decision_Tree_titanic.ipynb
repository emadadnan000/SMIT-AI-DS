# Decision Tree on Titanic Dataset
The dataset can be found here: https://www.kaggle.com/competitions/titanic/data
## Objective: 
build a model to predict if person would survive or not, based on following features: \
Pclass, 
Sex,
Age,
Fare


### Calculate score of your model


```python

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df.head()
df = df[['Survived','Pclass','Sex','Age','Fare']]

# Encode Sex
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# Handle missing values
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())

X = df[['Pclass','Sex','Age','Fare']]
y = df["Survived"]

```

```python

print("Feature names:", X.columns.tolist())
print("Target names:", y.name)
```

```python
print(X.head())
```

```python
print(y.head())
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print("Training size:", len(X_train))
print("Testing size:", len(X_test))
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(max_depth=4, random_state=42)
model.fit(X_train, y_train)
```

```python
print("Accuracy:", model.score(X_test, y_test))
```

```python
plt.figure(figsize=(12,8))
plot_tree(model, feature_names=X.columns.tolist(), class_names=['Not Survived', 'Survived'], filled=True)
plt.show()
```
